```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot 
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create label for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title 
plt.title("PyBer Fare by Month")

# Add the legend 
plt.legend()
```

```python
# Skill drill 5.1.5
# Create plot 
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

# Create label for the x and y axes 
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add the legend 
plt.legend()
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Skill Drill 5.1.6
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
plt.gca().invert_yaxis()

# Label the axes and title
ax.set_xlabel('Fares($)')
ax.set_ylabel('Date')
ax.set_title("PyBer Fare by Month")

# Add the legend
ax.legend()
```

```python
# Skill Drill 5.1.7
plt.scatter(y_axis, x_axis, c="red", label='Chicago')
plt.gca().invert_yaxis()

plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")

plt.xlim(0,)
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Skill Drill 5.1.7 
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', s = [i * 5 for i in y_axis], edgecolor='black', alpha=0.8, linewidths=2, label='Boston')
plt.gca().invert_yaxis()

ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
ax.set_xlim(0,)

ax.legend()
```

```python
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
# Skill Drill 5.1.8
fig, ax = plt.subplots(figsize=(8,8))

explode_values_sk = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels=x_axis,
       autopct='%.1f%%',
       colors=colors,
       explode=explode_values_sk,
       startangle=90,
       counterclock=False,
      shadow=True)
plt.show()
```
